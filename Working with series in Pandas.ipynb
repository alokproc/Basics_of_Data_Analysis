### Intro to Methods

Unlike attributes which present what is already given methods perform some operation on the data.

Methods do require parenthesis at the end.

```python
ratings = [7.7, 8.2, 8,9]
r = pd.Series(ratings)
r
```

```python
r.sum()
```

```python
r.product()
```

```python
r.mean()
```

### Parameters and Arguments

Parameters takes values which are known as arguments. For example if we set volume to 8 then 'Volume' is parameter and '8' is argument.


```python
titans = ["Robin", "BlueBeetle","Raven","BeastBoy","StarFire"]
pd.Series(titans)
powers = ['SuperStrength','SuperAgility','Flying','SuperHearing','Laser']
p = pd.Series(powers)
```

pd.Series takes first two arguments data and index. They can be given without mentioning parameters by giving them in order

```python
pd.Series(titans, powers)
```

or they can be given with parameter names to avoid ordering.

```python
pd.Series(index = powers, data = titans)
```

Key values can be repeated but they may limit some methods to be used.

```python
powers = ['SuperStrength','SuperAgility','Flying','SuperHearing','Flying']
p = pd.Series(powers)
```

```python
pd.Series(titans, powers)
```

### Inporting csv files with pandas

```python
pd.read_csv("pokemon.csv")
```

We need a series 'usecols' can be given as list as we can import more than one columns that we want. Pandas import a csv file as dataframe by default so 'squeeze' is used to import as 'Series'

```python
poke = pd.read_csv("pokemon.csv", usecols = ["Pokemon"], squeeze = True)
poke
```

```python
poke.head(6)
```

gives a summarised view from the top

```python
poke.tail(3)
```

gives a summarised view from the bottom.

```python
stockprice = pd.read_csv("google_stock_price.csv", usecols = ["Stock Price"], squeeze = True)
stockprice
```

```python
stockprice.head()
```

```python
stockprice.tail()
```
